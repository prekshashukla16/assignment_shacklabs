```python
import pandas as pd
import re
import numpy as np
from thefuzz import fuzz, process
```

## Importing the datasets and converting them to dataframes 

```python
df = pd.read_csv('amz_com-ecommerce_sample.csv',encoding='latin1')
df1 = pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding='latin1')
```

## Replacing the Varchar Unique Id with numerical Id 

```python
df = df.drop('uniq_id', axis = 1)
df1 = df1.drop('uniq_id', axis = 1)
```

```python
df['uniq_id'] = np.arange(len(df))
df1['uniq_id'] = np.arange(len(df1))
```

```python
df = df.set_index(df['uniq_id'])
df1 = df1.set_index(df1['uniq_id'])
```

## Dropping all the duplicates if there are any

```python
df.drop_duplicates()
```

```python
df1.drop_duplicates()
```

## Here the input of the product that needs to be searched for is to be given

```python
string = input("Enter to search:")
```

## The fuzz library is used to find the products similar to the ones user is looking for, the limit is set as 10 but it can be changed according to the preference

```python
df_amazon = pd.DataFrame(process.extract(string, df['product_name'], limit = 25, scorer=fuzz.token_sort_ratio), columns = ["product_name", "match score", "index"])
```

```python
df_amazon
```

```python
df_flipkart = pd.DataFrame(process.extract(string, df1['product_name'], limit = 25, scorer=fuzz.token_sort_ratio), columns = ["product_name", "match score", "index"])
```

```python
df_flipkart
```

## The indexes of all the records matching the product are extracted

```python
index_amazon = df_amazon['index']
index_amazon
```

```python
index_flipkart = df_flipkart['index']
index_flipkart
```

```python
df.columns
```

## The required details as metioned in the pdf are extracted for the matching products using indexes and the column names are also modified for ease

```python
df_output = df.loc[index_amazon]
df_output = df_output.rename(columns={'product_name':'amazon_product_name', 'retail_price':'amazon_retail_price',
                                      'discounted_price':'amazon_discounted_price'})
df_output[['amazon_product_name', 'amazon_retail_price', 'amazon_discounted_price']]
```

```python
df_output1 = df.loc[index_flipkart]
df_output1 = df_output1.rename(columns={'product_name':'flipkart_product_name', 'retail_price':'flipkart_retail_price',
                                      'discounted_price':'flipkart_discounted_price'})
df_output1[['flipkart_product_name', 'flipkart_retail_price', 'flipkart_discounted_price']]
```

## The final output of the product search is displayed as mentioned in the pdf for comparision by merging the dataframes

```python
final_result = pd.concat([df_output[['amazon_product_name', 'amazon_retail_price', 'amazon_discounted_price']],
                          df_output1[['flipkart_product_name', 'flipkart_retail_price', 'flipkart_discounted_price']]],
                        axis = 1)
final_result
```
